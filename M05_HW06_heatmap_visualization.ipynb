#### Problem 4
Since your instructor's gaming interests are pretty broad, you can help her decide to use the IGN scores to determine the choice of genre and platform.

1. Create a heatmap using the IGN scores by genre and platform and include the scores in the cells of the heatmap.
2. Base on the heatmap, which combination of genre and platform receives the highest average ratings?  Which combination receives the lowest average rankings? Write the answers in a markdown cell.

```python
## Create a heatmap
#lets break down the question so I can undestand it better
#1. heatmpap of scores by genre and platform
#okay I understand, we need to make the Platform column into the actical index of the dataframe, using .set_index()
platform_index = ign_scores.set_index("Platform")

plt.figure(figsize=(15, 10))
sns.heatmap(platform_index, annot=True, cmap="viridis") #annot to match positions
plt.title("IGN Scores by Platform and Genre")
plt.tight_layout()
plt.show()


#2. highekst and lowest average ranks for each combiination
#because the we have a 2D dataframe, we need to use .stack() to convert to 1D series to then apply idxmax and idxmin
print("Highest:", platform_index.stack().idxmax())
print("Lowest:", platform_index.stack().idxmin())
```

Highest: Playstation 4 & Simulation at 9.2 (highest by a big margin too)

Lowest: Game Boy Color & Fighting/Shooting

## Problem 5
Use the Pokemon dataset to create a clustermap with color. First, filter the dataframe to only keep data with `Type 1` equals to following values: `Water`, `Normal`, `Grass`, `Bug` and `Psychic`. Annotate the dendrogram using different colors for these five different `Type 1` values. Use `Name` as the index.

pokemon_data is in https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv

```python
# Problem 5
pokemon = pd.read_csv("https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv")
pokemon.head()
```

```python
#filter all of the types
types = ["Water", "Normal", "Grass", "Bug", "Psychic"]
#select only rows where Type 1 is in list of types
#and set index to name of pokemon
filtered = pokemon[pokemon["Type 1"].isin(types)].set_index("Name")
#map to colors
color_map = {"Water": "blue", "Normal": "gray", 
             "Grass": "green", "Bug": "brown", "Psychic": "pink"}
#map the associated colors to the type 1 column
row_colors = filtered["Type 1"].map(color_map)
#keep the numeric columns 
features = filtered.select_dtypes(include="number")
sns.clustermap(features, row_colors=row_colors, cmap="viridis", figsize=(12, 12), standard_scale=1)
plt.show()
```

This heatmap above uses the five different pokemon types with name as the index. The bright yellow represents a very high value for that specific type relative to others in the same column and the dark purple shows a low value comparison. This whole heatmap is entirely relative to all others in a column

Ex. Look at Snorlax on the bottom of the heatmap, you can see it is bright yellow for Total and Attach, telling us they have among the highest values for that stat across all the pokemon
